```python
import pandas as pd
import numpy as np
import nltk
from textblob import TextBlob
import re
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
data = {
    "tweets":[
        "I love the new iPhone! The camera quality is amazing.",
        "This phone battery is terrible. Very disappointed.",
        "The update is okay, nothing special.",
        "Amazing performance and smooth experience.",
        "Worst phone I have ever used."
    ]
}

df = pd.DataFrame(data)
df
```

```python
def clean_text(text):

    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#", "", text)
    text = re.sub(r"[^A-Za-z\s]", "", text)

    text = text.lower()

    return text

df["clean_tweets"] = df["tweets"].apply(clean_text)
df
```

```python
def get_sentiment(text):

    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:
        return "Positive"

    elif analysis.sentiment.polarity < 0:
        return "Negative"

    else:
        return "Neutral"

df["Sentiment"] = df["clean_tweets"].apply(get_sentiment)

df
```

```python
df["Sentiment"].value_counts()
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
import nltk
nltk.download('punkt_tab')
text = """
Artificial Intelligence is transforming the world rapidly. It is used in healthcare, education, finance,
transportation and many other industries. Machine learning and deep learning are subfields of AI that
help computers learn from data and make intelligent decisions. With the advancement of technology,
AI is expected to play a crucial role in solving complex real-world problems.
"""

parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()

summary = summarizer(parser.document, 2)

for sentence in summary:
    print(sentence)
```

```python
all_text = " ".join(df["clean_tweets"])

parser = PlaintextParser.from_string(all_text, Tokenizer("english"))
summarizer = LsaSummarizer()

summary = summarizer(parser.document, 2)

print("Summary of Tweets:\n")

for sentence in summary:
    print(sentence)
```
