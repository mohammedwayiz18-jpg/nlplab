```python
para = """Cricket is a bat-and-ball game that is played between two teams of eleven players on a field, at the centre of which is a 22-yard (20-metre; 66-foot) pitch with a wicket at each end, each comprising two bails (small sticks) balanced on three stumps. Two players from the batting team, the striker and nonstriker, stand in front of either wicket holding bats, while one player from the fielding team, the bowler, bowls the ball toward the striker's wicket from the opposite end of the pitch. The striker's goal is to hit the bowled ball with the bat and then switch places with the nonstriker, with the batting team scoring one run for each of these swaps. Runs are also scored when the ball reaches the boundary of the field or when the ball is bowled illegally."""
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
sentence = nltk.sent_tokenize(para)
```

```python
para1 = """Cricket is a bat-and-ball game that is played between two teams of eleven players on a field, at the centre of which is a 22-yard (20-metre; 66-foot) pitch with a wicket at each end, each comprising two bails (small sticks) balanced on three stumps. Two players from the batting team, the striker and nonstriker, stand in front of either wicket holding bats, while one player from the fielding team, the bowler, bowls the ball toward the striker's wicket from the opposite end of the pitch. The striker's goal is to hit the bowled ball with the bat and then switch places with the nonstriker, with the batting team scoring one run for each of these swaps. Runs are also scored when the ball reaches the boundary of the field or when the ball is bowled illegally."""

```

```python
sentence1 = nltk.sent_tokenize(para1)
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
for i in range(len(sentence1)):
  words = nltk.word_tokenize(sentence1[i])
  newword = [lemmatizer.lemmatize(word)for word in words]
  sentence1[i] = ' '.join(newword)
```
