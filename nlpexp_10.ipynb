```python
import nltk
import os
import re
import tarfile
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
import tarfile

tar = tarfile.open("review_polarity.tar.gz")
tar.extractall()
tar.close()

print("Dataset extracted successfully")
```

```python
data = []

pos_path = "txt_sentoken/pos"
neg_path = "txt_sentoken/neg"

for folder in [pos_path, neg_path]:
    for file in os.listdir(folder):
        with open(os.path.join(folder, file), encoding="latin-1") as f:
            data.append(f.read())

print("Total documents:", len(data))
```

```python
clean_text = []

for text in data:
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    clean_text.append(text)

print(clean_text[0][:500])
```

```python
import nltk
nltk.download('punkt_tab')
sentences = []

for text in clean_text:
    tokens = word_tokenize(text)
    sentences.append(tokens)

print(sentences[0][:20])
```

```python
model = Word2Vec(sentences,
                 vector_size=100,
                 window=5,
                 min_count=2,
                 workers=4)

print("Model trained successfully")
```

```python
words = model.wv.index_to_key

print("Vocabulary words:")
print(words[:20])
```

```python
vector = model.wv['movie']

print(vector)
```

```python
similar = model.wv.most_similar('movie')

print(similar)
```

```python
model.save("word2vec_movie_model.model")

print("Model saved successfully")
```
